```python
import requests
from bs4 import BeautifulSoup
```

```python
class pokemon:
    id = 0
    nome = "MissigNo."
    tipo_1 = "Normal"
    tipo_2 = ""
    
    def __init__(self, id, nome, tipo_1, tipo_2):
        self.id = id
        self.nome = nome
        self.tipo_1 = tipo_1
        self.tipo_2 = tipo_2

    def __str__(self):
        tipos = [self.tipo_1]
        if self.tipo_2 != "":
            tipos.append(self.tipo_2)
        string_formatada = f"No. {self.id}\n{self.nome}\n{' | '.join(tipos)}\n"
        return string_formatada
```

```python
def extrair_dados_pokemon_div(div_pokemon):
    attr_soup = div_pokemon.find_all("small")
    itype_soup = attr_soup[1].find_all(attrs={"class": "itype"})

    id = attr_soup[0].text.replace("#","")
    nome = div_pokemon.find_all(attrs={"class":"ent-name"})[0].text
    tipo_1 = itype_soup[0].text
    tipo_2 = ""
    if len(itype_soup) > 1:
        tipo_2 = itype_soup[1].text

    temp_pokemon = pokemon(id, nome, tipo_1, tipo_2)
    return temp_pokemon
```

```python
url = "https://pokemondb.net/pokedex/game/red-blue-yellow"
r = requests.get(url)
r.status_code
```

```python
page = r.text
soup = BeautifulSoup(page, 'html.parser')
pagina_principal = soup.find("main")
```

```python
infocard_list = pagina_principal.find_all('div', attrs={"class":"infocard-list infocard-list-pkmn-lg"})[0]
divs_pokemon = infocard_list.find_all('div', attrs={"class": "infocard"})
```

```python
pokedex = [extrair_dados_pokemon_div(pokemon) for pokemon in divs_pokemon]
```
